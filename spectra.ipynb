```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 24
tau = 3
def func(t, a=3, b=1):
    w0 = np.pi * 2 / T 
    w1 = np.pi * 2 / tau
    return ((a - b) / 2 * np.sin(t * w0) + (a + b) / 2) * np.sin(w1 * t)
```

```python
tv = np.linspace(0, 8 * T, 1600)
```

```python
plt.plot(tv, func(tv))
plt.grid()
```

```python
x = func(tv)
y = np.abs(np.fft.rfft(x))
freq = np.fft.rfftfreq(len(x))
```

```python
plt.plot(freq, y)
plt.grid()
plt.xlim(0, 0.08)
```
