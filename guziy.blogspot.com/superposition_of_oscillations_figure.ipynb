```python

"""
Code used to plot the graph in the problem statement.
"""

import matplotlib.pyplot as plt
import numpy as np

T = 24
tau = 4
W = 2 * np.pi / T
w0 = 2 * np.pi / tau

tv = np.linspace(0, T * 3, 5000)

a = 8
b = 2

xv = (0.5 * (a - b) * np.sin(W * tv) + 0.5 * (a + b)) * np.sin(w0 * tv)

ap = 0.5 * (a - b) * np.sin(W * tv) + 0.5 * (a + b)

plt.figure(figsize=(10, 6))
plt.rcParams["font.size"] = 22
plt.plot(tv, xv, linewidth=1.5)
plt.grid()
plt.plot(plt.xlim(), [0, 0], "k")

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(tv, ap, linestyle="dashed", color="k", linewidth=0.6)
plt.plot(tv, -ap, linestyle="dashed", color="k", linewidth=0.6)

plt.xticks([])
plt.yticks([])

# W * t = pi / 2 => ap=a
t1 = np.pi / (2 * W)
ax.annotate(text=r"$A$", xy=(t1, 0), xytext=(t1, a), ha="center", va="bottom", 
            arrowprops=dict(arrowstyle='<->', shrinkA=0, shrinkB=0, linewidth=2, alpha=0.5),
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0))

# W * t = 3 * pi / 2 => ap=b
t1 = 3 * np.pi / (2 * W)
ax.annotate(text=r"$B$", xy=(t1, 0), xytext=(t1, b), ha="center", va="bottom", 
            arrowprops=dict(arrowstyle="<->", shrinkA=0, shrinkB=0, linewidth=2, alpha=0.5),
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0))


# show periods

# W(t2 - t1) = 2pi
t1 = np.pi / (2 * W) + T
t2 = t1 + T
offset = 0.1 * a
ax.annotate(text=r"", xy=(t1, a + offset), xytext=(t2, a + offset), ha="left", va="center", 
            arrowprops=dict(arrowstyle="<->", shrinkA=0, shrinkB=0, linewidth=2, alpha=0.5),
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0))

ax.annotate(text=r"$T$", xy=(0.5 * (t1 + t2), a + offset), 
            xytext=(0.5 * (t1 + t2), a + offset), va="bottom", ha="center",
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0.2))

ax.plot([t1, t1], [a, a + 2 * offset], alpha=0.5, color="k")
ax.plot([t2, t2], [a, a + 2 * offset], alpha=0.5, color="k")


# w0 * (t2 - t1) = 2pi
t1 = 10 * tau
t2 = t1 + tau

v1 = (0.5 * (a - b) * np.sin(W * t1) + 0.5 * (a + b)) * np.sin(w0 * t1)
v2 = (0.5 * (a - b) * np.sin(W * t2) + 0.5 * (a + b)) * np.sin(w0 * t2)

offset = -a

ax.annotate(text=r"", xy=(t1, v1 + offset), xytext=(t2, v2 + offset), ha="left", va="center", 
            arrowprops=dict(arrowstyle="<->", shrinkA=0, shrinkB=0, linewidth=2, alpha=0.5),
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0))

ax.plot([t1, t1], [0, v1 + offset], alpha=0.5, color="k")
ax.plot([t2, t2], [0, v2 + offset], alpha=0.5, color="k")

ax.annotate(text=r"$\tau$", xy=(0.5 * (t1 + t2), a + offset), 
            xytext=(0.5 * (t1 + t2), offset * 0.95), va="bottom", ha="center",
            bbox=dict(boxstyle="square", fc="none", ec="none", pad=0.2))

```
